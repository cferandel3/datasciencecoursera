```python
def type1(x):
    a = pd.crosstab(x, y, margins=True)
    a['Non-Engaged'] = a['No']
    a['Engaged'] = a['Yes']
    a['% Vertical (Non-Engaged)'] = (a['No']*2/a['No'].sum()).round(2)
    a['% Vertical (Engaged)'] = (a['Yes']*2/a['Yes'].sum()).round(2)
    a = a.drop('No',axis=1)
    a = a.drop('Yes',axis=1)
    a = a.drop('All', axis = 1)
    a = a.fillna(0)
    a.to_clipboard(excel=True)
    return(a)
    
```

```python
def type4(x):
    a = pd.crosstab(x, y, margins=True)
    a['Non-Engaged'] = a['No']
    a['Engaged'] = a['No']*0
    a['% Vertical (Non-Engaged)'] = (a['No']*2/a['No'].sum()).round(2)
    a['% Vertical (Engaged)'] = a['No']*0
    a = a.drop('No',axis=1)
    a = a.drop('All', axis = 1)
    a = a.fillna(0)
    a.to_clipboard(excel=True)
    return(a)
```

```python
def type2(x):
    a = pd.crosstab(x, y, margins=False)
    return(a)
```

```python
def type3(a,b):
    c = pd.crosstab(a,b, margins=False)
    return(c)
```

```python
def concatType1(aList):
    group = []
    keyGroup = []
    for element in aList:
        group.append(type1(element[0]))
        keyGroup.append(element[1])
    
    result = pd.concat(group, keys = keyGroup, axis = 1)
    
    result = result.fillna(0)
    result.to_clipboard(excel=True)
    return(result)
    
    

# concatType1([(data.q6_DiscoveryCove,'DiscoveryCove'),(data.q6_SeaWorld, 'SeaWorld'),
#              (data.q6_Disney, 'Disney'),(data.q6_Universal,'Universal')])
```

```python
def concatType4(aList):
    group = []
    keyGroup = []
    for element in aList:
        group.append(type4(element[0]))
        keyGroup.append(element[1])
    
    result = pd.concat(group, keys = keyGroup, axis = 1)
    
    result = result.fillna(0)
    result.to_clipboard(excel=True)
    return(result)
    
```

```python
def breakdown(x, aList):
    b = len(aList)
    x = x.append(x.iloc[0:2,:].sum(axis = 0),ignore_index=True)
    x = x.append(x.iloc[b-3:b-1,:].sum(axis = 0),ignore_index=True)

    aDict = {}
    for y in range(b):
        aDict[y] = aList[y]
    aDict[b] = "Top 2"
    aDict[b+1] = "Bottom 2"
    x = x.rename(index = aDict)

    newList = [aList[0], aList[1], "Top 2"] + aList[2:b-1] + ["Bottom 2", "All"]
    x = x.reindex(newList)
    x = x.fillna(0)

    x.to_clipboard(excel=True)
    return(x)

```

```python
def append(aList):
    result = type2(aList[0])
    for unit in range(1,len(aList)):
        result = result.append(type2(aList[unit]))
          
    result['Non-Engaged'] = result['No']
    result['Engaged'] = result['Yes']
    result = result.drop('No',axis=1)
    result = result.drop('Yes',axis=1)
    result = result.fillna(0)
    
    result.to_clipboard(excel=True)
    return(result)
    
    
# append([data.q10_dolphins, data.q10_fish, data.q10_rivers, 
#        data.q10_birds, data.q10_beach, data.q10_amenties, data.q10_none])
```

```python
def append2(aList,y):
    result = type3(aList[0],y)
    for unit in range(1,len(aList)):
        result = result.append(type3(aList[unit], y))
    result = result.fillna(0)
    
    result.to_clipboard(excel=True)
    return(result)
    
    
# append([data.q10_dolphins, data.q10_fish, data.q10_rivers, 
#        data.q10_birds, data.q10_beach, data.q10_amenties, data.q10_none])

```
